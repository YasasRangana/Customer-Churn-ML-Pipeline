# Data Preprocessing & Feature Engineering

## Load the relevant libraries

```python
import numpy as nm
import pandas as pd
```

## Load the data

```python
df = pd.read_csv("../Data/raw/WA_Fn-UseC_-Telco-Customer-Churn.csv")
df.head()
```

## change TotalCharges Data type

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

### chech missing values and Fill missing

```python
df['TotalCharges'].isnull().sum()
```

```python
df['TotalCharges'] = df['TotalCharges'].fillna(df['TotalCharges'].median())
```

## Encode target variable

```python
df['Churn'] = df['Churn'].map({'Yes': 1, 'No': 0})
```

```python
df['Churn'].value_counts(normalize=True)
```

### Drop ID Column

```python
df.drop('customerID', axis=1, inplace=True)
```

## Separate features & target

```python
X = df.drop('Churn', axis=1)
y = df['Churn']
```

## Identify column types

```python
num_cols = ['tenure', 'MonthlyCharges', 'TotalCharges']
cat_cols = [col for col in X.columns if col not in num_cols]
```

```python
num_cols, cat_cols
```

## Build preprocessing pipeline

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), num_cols),
        ('cat', OneHotEncoder(handle_unknown='ignore'), cat_cols)
    ]
)
```

## Preprocessing Strategy
* Numeric features were scaled using StandardScaler.
* Categorical features were one-hot encoded.
* Missing values in TotalCharges were imputed using the median.
* Customer identifier was removed to prevent data leakage.
